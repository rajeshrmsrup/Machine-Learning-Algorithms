### // Implementation of Linear Regression. //

```python
# Linear Regression
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4,5])
y = np.array([7,14,15,18,19])
n = len(x)

# Calculate Xmean, Ymean, Sxx, Sxy to find value of slop and intercept
x_mean = np.mean(x)
y_mean = np.mean(y)
print("Mean(x) =", x_mean)
print("Mean(y) =", y_mean)

Sxy = np.sum(x*y)-n*x_mean*y_mean
Sxx = np.sum(x*x)-n*x_mean*x_mean
b1 = Sxy/Sxx
b0 = y_mean-b1*x_mean
print("Slope b1 is", b1)
print("Intercept b0 is", b0)

#Data points and fit the regression line
y_pred = b1*x+b0

plt.scatter(x,y, color='red')
plt.plot(x,y_pred, color='green')
plt.xlabel("Independent variable x")
plt.ylabel("Dependent variable y")

# Analyze the performance of the modelby calculating mean squared error and R^2
error = y-y_pred
se = np.sum(error**2)
print("Squared error is", se)

mse = se/n
print('Mean squared error is', mse)

rmse = np.sqrt(mse)
print("Root square error is", rmse)

SSt = np.sum((y-y_mean)**2)
R2 = 1-(se/SSt) #deter coef
print("R sqare is", R2)
```
